```python
from impy.models import Pythia8
from impy.kinematics import CenterOfMass
from impy.constants import TeV, GeV, quarks_and_diquarks_and_gluons
import numpy as np

ekin = CenterOfMass(10 * GeV, "proton", "proton")

gen = Pythia8(ekin, 1)

events = list(ev.copy() for ev in gen(10))
```

```python
event = events[0]
event
```

```python
import pyhepmc

with pyhepmc.open("pythia8.dat", "w") as f:
    f.write(event)

with pyhepmc.open("pythia8.dat") as f:
    event3 = f.read()

event3
```
